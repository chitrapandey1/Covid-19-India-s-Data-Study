modules used

```python
import requests as req
import json
import pandas as pd
import csv
```

```python
url_data = 'https://data.covid19india.org/v4/min/data.min.json'
url_timeseries = 'https://data.covid19india.org/v4/min/timeseries.min.json'
```

Variable has been created of url string

##### Requesting Data from URL

```python
r_url_data = req.get(url_data)
r_url_time = req.get(url_timeseries)
```

#### Parsing Data

collecting the json data direct from url text method has been used


```python
data_json = r_url_data.text
timeseries_json = r_url_time.text
```

```python
data = json.loads(data_json)
data1 = json.loads(data_json)
timeseries = json.loads(timeseries_json)

```

```python
for x in data1:
    try:
        
        del data1[x]['districts']
    except:
        pass
```

we have created the different files for district data and state data that's why we have deleted the district from state table file

```python

timeseries_list = []
for x in timeseries:
    st = x
    for y in timeseries[x]['dates']:
        try:
            timeseries[x]['dates'][y]['state'] = st
            timeseries[x]['dates'][y]['date'] = y
            timeseries_list.append(timeseries[x]['dates'][y])
        except:
            pass
```

appending the state column in timeseriese file

timeseries list has been created which data for each day

```python
timeseries_final = pd.json_normalize(timeseries_list)
timeseries_final
```

```python
data_list =  [x for x in data1.values()]
state_list = [ x for x in data1]
for x in range(len(data_list)):
    data_list[x]['state'] = state_list[x]
```

```python
state_data = pd.json_normalize(data_list)   #without districts
```

```python
state_data
```

```python
data_list_1 =  [x for x in data.values()]
state_list_1 = [ x for x in data]
for x in range(len(data_list)):
    try:
        data_list_1[x]['districts']['state'] = state_list[x] 
    except:
        pass
```

```python
district_data = []
for x in data_list_1:
    try:
        district_data.append(x['districts'])
    except:
        pass
```

```python
for x in district_data:
    st = x['state']
    for y in x:
        try:
            x[y]['state'] = st
            x[y]['district'] = y
        except:
            pass
```

```python
district_data_1 = []
for x in district_data:
    for y in x:
        if y == 'state':
            continue
        
        district_data_1.append(x[y])
```

```python
district_data_final = pd.json_normalize(district_data_1)
```

```python
district_data_final
```

```python
district_data_final.to_csv('dsitrict_data.csv')
timeseries_final.to_csv('timeseries_data.csv')
state_data.to_csv('state_data.csv')
```
