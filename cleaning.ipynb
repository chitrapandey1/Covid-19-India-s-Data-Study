# Getting familiar with data

```python
import pandas as pd
import numpy as np
```

```python
file_name = 'dsitrict_data.csv'
```

```python
df = pd.read_csv(file_name)
```

```python
df.shape
```

```python
df.head()
```

```python
df.drop(df.columns[0],axis=1,inplace=True)
```

dropping first irrelavant column

Finding the coolumn with nan values

```python
df_nan = [x for x in df.columns if df[x].isnull().sum()>1]
```

```python
df_nan
```

Finding the how much % is nan values in particular column

```python
for x in df_nan:
    print(x,np.round(df[x].isnull().mean(),4)*100,'% missing values' )
```

Dropping the Columns which are not relevant % has been decided manually by looking at column name

```python
df.drop([x for x in df_nan if np.round(df[x].isnull().mean(),4)*100 > 80],axis=1,inplace=True)
```

```python
df.to_csv(file_name+'_filtered.csv')
```
